```python
import pdfplumber
from tabulate import tabulate

def extract_text_from_pdf(pdf_path):
    text_data = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                text_data.append(text)
    return text_data

def organize_text(text_data):
    organized_data = {}
    current_section = None

    for text in text_data:
        lines = text.split("\n")
        for line in lines:
            if line.strip():  # Se a linha não estiver vazia
                if ":" in line:  # Detecção de novas seções pelo caractere ":"
                    parts = line.split(":", 1)
                    current_section = parts[0].strip()
                    content = parts[1].strip()
                    if current_section not in organized_data:
                        organized_data[current_section] = []
                    organized_data[current_section].append(content)
                elif current_section:
                    organized_data[current_section].append(line.strip())

    return organized_data

def display_organized_text(organized_data):
    for section, content in organized_data.items():
        print(f"\n{section}:\n{'-'*len(section)}")
        print("\n".join(content))

pdf_path = "cv.pdf"
text_data = extract_text_from_pdf(pdf_path)
organized_data = organize_text(text_data)
display_organized_text(organized_data)
```

```python
import fitz  # PyMuPDF
import re

def extract_text_from_pdf(pdf_path):
    # Abrir o documento PDF
    document = fitz.open(pdf_path)
    
    # Extração de texto de todas as páginas
    text = ""
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text()
    
    return text

def organize_text(text):
    # Separar o texto em seções básicas
    sections = {
        "Nome": "",
        "Contato": "",
        "Educação": "",
        "Experiência Profissional": "",
        "Habilidades": ""
    }

    # Definir padrões para cada seção
    patterns = {
        "Nome": r"^(.*)\n",  # Nome na primeira linha
        "Contato": r"Contato:(.*)",  # Contato
        "Educação": r"Educação:(.*?)(?=Experiência Profissional|$)",  # Educação até Experiência Profissional ou fim
        "Experiência Profissional": r"Experiência Profissional:(.*?)(?=Habilidades|$)",  # Experiência Profissional até Habilidades ou fim
        "Habilidades": r"Habilidades:(.*)"  # Habilidades
    }
    
    for section, pattern in patterns.items():
        match = re.search(pattern, text, re.DOTALL | re.IGNORECASE)
        if match:
            sections[section] = match.group(1).strip()
    
    return sections

def main(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    organized_text = organize_text(text)
    
    # Exibir o texto organizado
    for section, content in organized_text.items():
        print(f"{section}:\n{content}\n{'-'*40}")

# Exemplo de uso
pdf_path = "cv.pdf"


text = extract_text_from_pdf(pdf_path)

```

```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    # Abrir o documento PDF
    document = fitz.open(pdf_path)
    
    # Extração de texto de todas as páginas
    text = ""
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text("text")
    
    return text

def organize_text_into_paragraphs(text):
    # Separar o texto em parágrafos usando linhas em branco como delimitadores
    paragraphs = text.split('\n\n')
    
    # Remover parágrafos vazios e limpar espaços extras
    paragraphs = [para.strip() for para in paragraphs if para.strip()]
    
    return paragraphs

def main(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    paragraphs = organize_text_into_paragraphs(text)
    
    # Exibir os parágrafos organizados
    for i, para in enumerate(paragraphs, 1):
        print(f"{i}:\n{para}\n{'-'*40}")

# Exemplo de uso
pdf_path = "cv.pdf"
main(pdf_path)
```

```python
def organize_resume_text(text):
    # Dicionário para armazenar as seções organizadas
    sections = {
        "Resumo Profissional": "",
        "Habilidades": "",
        "Qualificações": "",
        "Experiência": "",
        "Educação": "",
        "Histórico de Trabalho": ""
    }

    # Definir padrões para cada seção
    patterns = {
        "Resumo Profissional": r"Professional Summary(.*?)(?=\nSkills|\nCore Qualifications|\nExperience|\nEducation|\nWork History|\n$)",
        "Habilidades": r"Skills(.*?)(?=\nCore Qualifications|\nExperience|\nEducation|\nWork History|\n$)",
        "Qualificações": r"Core Qualifications(.*?)(?=\nExperience|\nEducation|\nWork History|\n$)",
        "Experiência": r"Experience(.*?)(?=\nEducation|\nWork History|\n$)",
        "Educação": r"Education(.*?)(?=\nWork History|\n$)",
        "Histórico de Trabalho": r"Work History(.*?)(?=$)"
    }

    # Extrair cada seção usando regex
    for section, pattern in patterns.items():
        match = re.search(pattern, text, re.DOTALL | re.IGNORECASE)
        if match:
            sections[section] = match.group(1).strip()

    return sections
```
